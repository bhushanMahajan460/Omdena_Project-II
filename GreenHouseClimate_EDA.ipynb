## **Parameters Description**

Tair - Greenhouse Air temperature

Rhair - Greenhouse relative humidity

CO2air - CO2 greenhouse

HumDef - Greenhouse humidity deficit

VentLee - Leeward vents opening

Ventwind - Windward vents opening

AssimLight - HPS lamps status (on-off)

EnScr - Energy curtain opening 

BlackScr - Blackout curtain opening

PipeLow - Rail pipe Temperature (Lower
circuit) 

PipeGrow - Crop pipe Temperature (Growth
circuit) 

co2_dos - CO2 dosing 

Tot_PAR - Total inside PAR
(Sun + HPS + LED)

Tot_PAR_Lamps - PAR sum from HPS and LED lamps
EC_drain_PC - Drain EC

pH_drain_PC - Drain pH

Water_sup - Cumulative number of minutes of
irrigation in a day

Cum_irr - Cumulative number of litres of
irrigation in a day

### Most of parameters have object type but they have numerical value so we convert them into float

```python
#convert object into float data type
for i in ghc:
  ghc[i] = ghc[i].astype(float)
```

```python
ghc.describe()
```

```python
ghc.shape
```

### Count null values in each parameter and remove that null entry row 

```python
ghc.isna().sum()
```

```python
ghc.dropna(axis=0, inplace=True)
```

```python
ghc.isna().sum()
```

### Find Correlation among each parameter and plot them

```python
correlation=ghc.corr()
correlation
```

```python
plt.rcParams['figure.figsize'] = [30, 24]
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns,annot=True,cmap='inferno')
plt.show()
```

# **Among all these parameters,**

*   **(AssimLight, CO2), (HumDef, Tair), (AssimLight, Tot_PAR_Lamps), (CO2, Tot_PAR_Lamps), (HumDef, Tair), (HumDef, Tot_PAR), (Tair, Tot_PAR),  (CO2, Tot_PAR), (VentLee, Ventwind) parameters have highly positive correlation.**

*   **And (Rhair, HumDef), (BlackScr, HumDef), (BlackScr, Tair), (BlackScr, Tot_PAR) parameters have highly negative correlation**



# Line Plot : Paramter vs Time plot

```python
for i in ghc:
  plt.figure(figsize=(14,6))
  sns.lineplot(data=ghc, x="%time", y=ghc[i])
  plt.title("{} vs Time".format(i))
  plt.xlabel("Time")
  plt.ylabel(i)
  plt.show()
```

## From all these unstable graph, It is clear that they dropped/increased these parameters level inside green house in day/night to maintain greenhouse atmosphere like day and night.

```python
for i in ghc:
  plt.figure(figsize=(14,6))
  sns.boxplot(x=ghc[i])
  plt.title("{} Box Plot".format(i))
  plt.show()
```

# **From the above boxplots, we can easily find minimum level, maximum level, and average level of each parameter inside greenhouse**
